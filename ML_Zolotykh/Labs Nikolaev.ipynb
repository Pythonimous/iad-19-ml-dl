Keywords, Overview, Tagline и Title все являются текстовыми характеристиками непосредственно фильма, их можно объединить в один текст и репрезентировать вместе (используя, например, предобученную на Wiki GLoVE модель). С другой стороны, production_companies надо представлять отдельно. Для этого попробуем найти в интернете текст про film production companies, соскрапим его и обучим отдельную модель самостоятельно. Этот шаг, чтобы не перегружать главный jupyter файл, я произвёл в отдельном файле 'Production Embeddings.ipynb'

Итак, мы обучили модель размерностью 50 и сохранили её в папке models (а заодно скачали модели glove с http://nlp.stanford.edu/data/glove.6B.zip, и будем использовать их для представления текстовой информации про сам фильм).

Сначала займёмся Production companies.

```python
prod_comps = [a.lower().replace('|', ' ') for a in data['production_companies']]
```

```python
prod_comps[:4]
```

```python
from gensim.models import doc2vec, KeyedVectors
from nltk.tokenize import word_tokenize
```

```python
prod_model = doc2vec.Doc2Vec.load('models/prod_companies.d2v')
```

```python
prod_vectors = [prod_model.infer_vector(word_tokenize(p)) for p in tqdm(prod_comps)]
```

```python
prod_vectors = pd.DataFrame(np.array(prod_vectors)).add_prefix('pv_')
prod_vectors.head()
```

Теперь Texts. Сначала преобразуем glove-txt формат в читаемый gensim'ом, затем загрузим. (используем 50d модель)

```python
from gensim.scripts.glove2word2vec import glove2word2vec

glove_input_file = 'models/glove.6B.50d.txt'
word2vec_output_file = 'models/glove.6B.50d.txt.word2vec'
glove2word2vec(glove_input_file, word2vec_output_file)
```

```python
text_model = KeyedVectors.load_word2vec_format(word2vec_output_file, binary=False)
```

Объединим тексты и подвергнем базовой предобработке.

```python
film_desc = data['title'] + ' ' + data ['tagline'] + ' ' + data['overview'] + ' ' + data['keywords']
```

```python
data.insert(0, 'film_desc', film_desc)
```

```python
data = data.drop(['title', 'tagline', 'overview', 'keywords'], axis=1)
```

```python
data['film_desc'].head()
```

Нижний регистр:

```python
data['film_desc'] = data['film_desc'].apply(lambda x: " ".join(x.lower() for x in x.split()))
```

Пунктуация:

```python
data['film_desc'] = data['film_desc'].str.replace('[^\w\s]','')
```

Стоп-слова:

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

```python
data['film_desc'] = data['film_desc'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```

```python
data['film_desc'].head()
```

```python
film_desc = data['film_desc']
```

Получим векторы описаний как среднее из векторов их слов.

```python
film_vecs = []
for d in tqdm(film_desc):
    words = word_tokenize(d)
    sen_vec = []
    
    for w in words:
        try:
            sen_vec.append(text_model[w])
        except KeyError:
            sen_vec.append(np.zeros((50,))) # Незнакомые слова заменяем нулевым вектором размерностью 50.
            
    film_vecs.append((sum(sen_vec)/len(sen_vec)))
```

```python
film_vecs = pd.DataFrame(np.array(film_vecs)).add_prefix('fv_')
film_vecs.head()
```

В условиях ограниченности данных, вероятнее всего (если всё же рассматривать эту задачу как классификационную, а не регрессионную), стоит объединить extremes в соответствующие группы (as-is это скорее выбросы).

## 7) Модели
### K-NN:

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

knn = KNeighborsClassifier(n_neighbors = 10)

knn.fit(X_train, y_train)
```

Подберём оптимальные значения соседей:

```python
from sklearn.model_selection import GridSearchCV
nnb = [1, 3, 5, 10, 15, 20, 25, 35, 45, 55]
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=10, verbose = 0)
grid.fit(X_train, y_train)

best_cv_err = 1 - grid.best_score_
best_n_neighbors = grid.best_estimator_.n_neighbors
print(best_cv_err, best_n_neighbors)
```

```python
knn = KNeighborsClassifier(n_neighbors = 1)

knn.fit(X_train, y_train)

y_test_predict = knn.predict(X_test)
err_test  = np.mean(y_test != y_test_predict)

err_test
```

```python
print(classification_report(y_test, y_test_predict))
```

## K-nearest, как правило, не слишком эффективен для задач с данными высокой размерности, а для NLP зачастую лучше работает Naive Bayes. Попробуем другие методы

Как мы видим, все векторные (pv, fv) и изначально численные признаки обладают высокой важностью относительно остальных. Этим же отличается и большая часть "жанровых" признаков (кроме Foreign).

Исключим все признаки с важностью меньше 0.001: это и уберёт малозначительные фичи, и более, чем в два раза, уменьшит размерность.

Напишем простую функцию для приведения списка к какой-то длине: если нужно расширить, она расширяет, если нужно сузить, она сужает.

```python
import random
```

```python
def sample_by_len(to_sample, length): # сэмплим список к какой-то длине:
                                      # oversampling или undersampling, в зависимости, больше или меньше нужно
    body_multi = length // len(to_sample) # сколько раз нам надо расширить оригинальный список
    tail_len = length % len(to_sample) # сколько ещё добавить
    sample = to_sample * body_multi + random.sample(to_sample, tail_len)
    
    return sample
```

### Undersampling

```python
less4_u = sample_by_len(less4, min_len)
five_u = sample_by_len(five, min_len)
six_u = sample_by_len(six, min_len)
seven_u = sample_by_len(seven, min_len)
great8_u = sample_by_len(great8, min_len)

total_u = less4_u + five_u + six_u + seven_u + great8_u # список индексов
```

```python
X_train_u = [X_train[i] for i in total_u]
y_train_full_u = sum([[i]*min_len for i in ['<=4', '5', '6', '7', '>=8']], [])
```

#### Shuffle_data: простая ф-ция для перемешивания данных и ярлыков в одной последовательности

```python
def shuffle_data(x, y):
    data_full = list(zip(x, y))
    random.shuffle(data_full)
    x, y = zip(*data_full)
    return x, y
```

```python
X_train_u, y_train_full_u = shuffle_data(X_train_u, y_train_full_u)
```

```python
X = data.drop('vote_average', axis = 1) # выбрасываем столбец 'class'
y = data['vote_average']
feature_names = X.columns
```

```python
X = X.to_numpy()
```

```python
from sklearn.preprocessing import LabelBinarizer
lb = LabelBinarizer()
```

```python
y = lb.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 666)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)

N, d = X.shape
```

```python
y_train = lb.inverse_transform(y_train)
y_test = lb.inverse_transform(y_test)
```

## Займёмся нейронной сетью

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp_model = MLPClassifier(hidden_layer_sizes = (100,), solver = 'lbfgs', 
                          activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)
```

```python
print(np.mean(y_train != y_train_pred), np.mean(y_test != y_test_pred))
```

```python
print(mlp_model.score(X_train, y_train), mlp_model.score(X_test, y_test))
```

## Не ахти как, попробуем для разных alpha и количества нейронов.

```python
alpha_arr = np.logspace(-3, 2, 21)
test_err = []
train_err = []
for alpha in tqdm(alpha_arr):
    mlp_model = MLPClassifier(alpha = alpha, hidden_layer_sizes = (100,), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
plt.semilogx(alpha_arr, train_err, 'b-o', label = 'train')
plt.semilogx(alpha_arr, test_err, 'r-o', label = 'test')
plt.xlim([np.max(alpha_arr), np.min(alpha_arr)])
plt.title('Error vs. alpha')
plt.xlabel('alpha')
plt.ylabel('error')
plt.legend()
pass
```

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

```python
alpha_opt = alpha_arr[test_err == min_test_err] # optimal alpha
alpha_opt = alpha_opt[0]
print(alpha_opt)
```

Пробуем!

```python
mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = (100,),
                          solver = 'lbfgs', activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)

y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
print(np.mean(y_train != y_train_pred))
print(np.mean(y_test != y_test_pred))
```

```python
print(mlp_model.score(X_train, y_train), mlp_model.score(X_test, y_test))
```

## Мы видим ухудшение результатов на тренировочном корпусе при улучшении на тестовом (3%). Теперь покалибруем число нейронов (для 1-2 слоёв)

```python
neurs = [(i,) for i in range(50, 401, 50)] + [(i,z,) for i in range(50,201,50) for z in range(50,201,50)]
test_err = []
train_err = []
for n in tqdm(neurs):
    mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = n, 
                              solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
neurs_label = [str(i) for i in neurs]
```

```python
plt.plot(neurs_label[:8], train_err[:8], 'b-o', label = 'train')
plt.plot(neurs_label[:8], test_err[:8], 'r-o', label = 'test')
plt.title('Error vs. neurons')
plt.xlabel('neurons')
plt.ylabel('error')
plt.legend()
pass
```

```python
plt.plot(neurs_label[8:16], train_err[8:16], 'b-o', label = 'train')
plt.plot(neurs_label[8:16], test_err[8:16], 'r-o', label = 'test')
plt.title('Error vs. neurons')
plt.xlabel('neurons')
plt.ylabel('error')
plt.legend()
pass
```

```python
plt.plot(neurs_label[16:24], train_err[16:24], 'b-o', label = 'train')
plt.plot(neurs_label[16:24], test_err[16:24], 'r-o', label = 'test')
plt.title('Error vs. neurons')
plt.xlabel('neurons')
plt.ylabel('error')
plt.legend()
pass
```

<b> Графики в этом блоке выглядят странно. </b> Это объясняется тем, что по оси абсцисс отложены конфигурации на разное количество нейронов. На первом - (50, x) и (100, x), на втором - (150, x) и (200, x) По сути, четыре графика, свёрнутые в два. Безусловно, не самое удачное дизайнерское решение, но позволяет увидеть, при каком количестве нейронов результат оптимален.

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

```python
neur_opt = neurs[test_err.index(min_test_err)] # optimal neurons
print(neur_opt)
```

```python
mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = neur_opt,
                          solver = 'lbfgs', activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)

y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
print(np.mean(y_train != y_train_pred))
print(np.mean(y_test != y_test_pred))
```

## By default, 100 estimators. Let's try different numbers.

```python
tree_nums = list(range(100, 701, 50))
test_err = []
train_err = []
oob_err = []
for trees in tqdm(tree_nums):
    clf = RandomForestClassifier(n_estimators = trees, max_depth=5, random_state=0, oob_score = True)
    clf.fit(X_train, y_train)

    y_train_pred = clf.predict(X_train)
    y_test_pred = clf.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
    
    oob_err.append(1 - clf.oob_score_)    
```

```python
plt.semilogx(tree_nums, train_err, 'b-o', label = 'train')
plt.semilogx(tree_nums, test_err, 'r-o', label = 'test')
plt.semilogx(tree_nums, oob_err, 'g-o', label = 'out of bag')
plt.xlim([min(tree_nums), max(tree_nums)])
plt.title('Error vs. trees')
plt.xlabel('trees')
plt.ylabel('error')
plt.legend()
pass
```

## Проверим по разным метрикам

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score,\
                            homogeneity_score, completeness_score, v_measure_score,\
                            adjusted_rand_score, adjusted_mutual_info_score, accuracy_score
```

<b> Unsupervised metrics: </b>
    - Silhouette - how similar object is to its own cluster compared to other clusters;
    - D-H - ratio between within-cluster and between-cluster dispersion (Variance Ratio Criterion);
    - D-B - average similarity measure of each cluster with its most similar

```python
def scores(name, x, y_true, classifier):
    silh = silhouette_score(x, classifier.labels_)  # 1.0 is perfect
    cal_har = calinski_harabasz_score(x, classifier.labels_)  # Higher is better
    dav_bou = davies_bouldin_score(x, classifier.labels_)  # 0.0 is perfect
    homo = homogeneity_score(y_true, classifier.labels_)  # 1.0 is perfect
    compl = completeness_score(y_true, classifier.labels_)  # 1.0 is perfect
    v_meas = v_measure_score(y_true, classifier.labels_)  # 1.0 is perfect
    adj_rand = adjusted_rand_score(y_true, classifier.labels_)  # 1.0 is perfect
    adj_mutual = adjusted_mutual_info_score(y_true, classifier.labels_,\
                                            average_method='arithmetic')  # 1.0 is perfect
    error = 1 - accuracy_score(y_true, classifier.labels_)  # Lower is better
    return [[name, silh, cal_har, dav_bou, homo, compl, v_meas, adj_rand, adj_mutual, error]]
```

```python
columns_list = ['Algorithm', 'Silh', 'Cal-Hb', 'Dav-Bdn', 'Homo', 'Comp', 'Vm', 'A-Rand', 'A-Mut', 'Error']
```

```python
scores_table = pd.DataFrame(columns = columns_list)
```

Для сравнения результатов нас интересуют только алгоритмы, которые предварительно дали нормальное количество кластеров (5; остальные даже смысла не имеет смотреть). Здесь это K-means (оба) и Agglomerative Clustering. Остальные видно, что провалились уже на первом этапе отбора.

```python
kmeans_rand_result = pd.DataFrame(scores('K-means Random', X, y, kmeans_rand), columns = columns_list)
scores_table = scores_table.append(kmeans_rand_result)
```

```python
kmeans_plus_result = pd.DataFrame(scores('K-means Plus',  X, y, kmeans_plus), columns = columns_list)
scores_table = scores_table.append(kmeans_plus_result)
```

```python
ac_result = pd.DataFrame(scores('Agglomerative Clustering',  X, y, ac), columns = columns_list)
scores_table = scores_table.append(ac_result)
```

```python
scores_table = scores_table.reset_index(drop=True)
```

```python
scores_table
```

Поскольку у Silhouette наилучшее значение 1, and а худшее -1,  кластеризаторы соглашаются, что кластеры пересекаются. По supervised метрикам же результаты очень плохие, и близкая к 1.0 показывает по большей части некорректное разбиение.
<br>
<br>
Это объяснимо, потому что в пусть скалированных, но разноплановых (эмбеддинги, популярность, временные метки, средняя оценка, и т.д.) данных такой размерности закономерности абсолютно неочевидны, и классификаторы не способны их выделить правильно.

Верное количество кластеров выделяется только там, где оно поступает на вход как параметр. При бесконтрольном же количестве, их получается до нескольких сотен.

K-means кластеры получаются более гомогенные внутри, нежели у Agglomerative Clustering (по всем показателям). Однако это касается K-means со случайным выбором центров.
